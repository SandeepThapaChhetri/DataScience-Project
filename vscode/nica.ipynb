```python
#importing the pandas library
import pandas as pd
```

```python
#reading the file
df = pd.read_csv('../data-companywise/NICA.csv')
df
```

```python
# Rename specific columns to the "Perfect Schema"
df = df.rename(columns={
    'published_date': 'date',
    'traded_quantity': 'volume',
    'traded_amount': 'turnover',
    'per_change': 'price_change_pct'
})
```

```python
df
```

```python
#year column created
df['date'] = pd.to_datetime(df['date'])   # convert first
df['year'] = df['date'].dt.year
```

```python
df
```

```python
#filtering the data from 2022
df['date'] = pd.to_datetime(df['date'])
df=df[df['date'].dt.year.between(2022,2025)].copy()
```

```python
df
```

```python
#resetting index
df.reset_index(drop= True, inplace = True)
df
```

```python
fundamental_data = {
    'year': [2022, 2023, 2024, 2025], # Corresponding to FY 78/79, 79/80, 80/81, 81/82
    'eps': [39.13, 39.65, 9.26, 1.08],
    'pe-ratio': [18.10, 19.80, 48.80, 321.90],
    'roe': [19.54, 17.58, 4.57, 0.54],
    'dividend': [0.00, 30.53, 0.00, 0.00]
}
df_fundamental = pd.DataFrame(fundamental_data)
df_fundamental
```

```python
# merging
df = pd.merge(df, df_fundamental, on='year',how= 'left')
df
```

```python
df = df.copy()
df.to_csv('../data-companywise/NICA.csv', index = False)
```
